```python
def factorial(n):     
    # single line to find factorial
    return 1 if (n==1 or n==0) else n * factorial(n - 1);
```

```python
import numpy as np
x = np.array([1, 3, 5, 7, 8, 9, 120])
```

```python
factorial(x)
```

```python
np.math.factorial(4)
```
