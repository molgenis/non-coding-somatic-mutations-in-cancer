```python
import pandas as pd
import numpy as np
from statsmodels.sandbox.stats.multicomp import multipletests
```

```python
path_file = "D:/Hanze_Groningen/STAGE/analyse/new/UCNE_NonCoding_both_0_TESTS.tsv"
df = pd.read_csv(path_file, sep='\t')
df = df[['chr', 'start_position_regio', 'end_position_regio', 'counts_breast', 'counts_nonbreast',
                    'p_value_X2_self', 'p_value_X2', 'p_value_F']]
```

```python
def run_multiple_testing_correction(df, alpha, method, type_test):
    list_p_values = list(df[f'p_value_{type_test}'])
    rej, p_adjusted = multipletests(list_p_values, alpha=alpha, method=method)[:2]
    df[f'{method}_{type_test}'] = rej
    return df
```

```python
alpha=0.05
method = 'bonferroni'
df = run_multiple_testing_correction(df, alpha, method, 'X2_self')
# df = run_multiple_testing_correction(df, alpha, method, 'X2')
# df = run_multiple_testing_correction(df, alpha, method, 'F')
# print(set(df['p_value_X2']))
# print(set(df['bonferroni_X2_self']))
# df[df['bonferroni_X2_self'] != 1.0][['bonferroni_X2_self', 'p_value_X2']].sample(30)

df['bonferroni_X2_self'].value_counts()
df['bonferroni_test'] = df['p_value_X2'] < (0.05/len(set(df)))
# df
```
