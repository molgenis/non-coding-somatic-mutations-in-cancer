```python
# import required module
import os
import itertools
from collections import Counter

# assign directory
directory = 'E:/analyse/new/correction/genes/bon/all/'

dict_test = dict()
all_list = list()
breast_list = list()
nonbreast_list = list()
# iterate over files in
# that directory
for filename in os.listdir(directory):
    print(filename)
    type_layer = filename.split('_')[0]
    f = os.path.join(directory, filename)
    df = pd.read_csv(f, sep='\t')
    print(len(df))
    dict_test[filename] = len(df)
    genes = list(itertools.chain(*df['hg19.kgXref.geneSymbol'].str.split(",").tolist()))
    if len(genes) > 1:
        if '_ALL_close_gene' in filename:
            all_list.append(genes)            
        elif '_breast_close_gene' in filename:
            breast_list.append(genes)            
        elif '_nonbreast_close_gene' in filename:
            nonbreast_list.append(genes)
        else:
            print(f'ERROR: {filename}')   

all_list = list(itertools.chain(*all_list))
breast_list = list(itertools.chain(*breast_list))
nonbreast_list = list(itertools.chain(*nonbreast_list))
all_dict = dict(sorted(dict(Counter(all_list)).items(), key=lambda item: item[1], reverse=True))
breast_dict = dict(sorted(dict(Counter(breast_list)).items(), key=lambda item: item[1], reverse=True))
nonbreast_dict = dict(sorted(dict(Counter(nonbreast_list)).items(), key=lambda item: item[1], reverse=True))
print(dict_test)
```

```python
path = "C:/Users/Anne_/Downloads/cosmic_genes.csv"
df_cosmic = pd.read_csv(path, sep=';')
cosmic_genes = list(df_cosmic['Gene Symbol'])
tumor_types = list(df_cosmic['Tumour Types(Somatic)'])
```

```python
list_column_names = ['ALL', 'breast', 'nonbreast']
for index, select_dict in enumerate([all_dict, breast_dict, nonbreast_dict]):
    key_list = list(select_dict.keys())
    value_list = list(select_dict.values())
    gene_df = pd.DataFrame(columns=['gene_name', 'count'])
    gene_df['gene_name'] = key_list
    gene_df['count'] = value_list
    in_cosmic = list()
    for gene in key_list:
        if gene in cosmic_genes:
            print(gene)
            in_cosmic.append(True)
        else:
            in_cosmic.append(False)
    gene_df['cosmic_gene'] = in_cosmic
    print(gene_df)
    path_new_file = f"E:/analyse/new/correction/genes/bon/all_list2/{list_column_names[index]}_genes.tsv"
    gene_df.to_csv(path_new_file, sep='\t', encoding='utf-8', index=False) 
    
```
