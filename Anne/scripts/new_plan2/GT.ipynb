# Get all data

```python
def get_project_ids(db):
    db.cursor.execute("""SELECT *
                            FROM project""")
    projects = db.cursor.fetchall()
    dict_project = dict()
    for proj in projects:
        dict_project[proj['ID']] = proj['project_ID']
    return dict_project
```

```python
path_db = 'D:/Hanze_Groningen/STAGE/db_laatste_copy.db' #'D:/Hanze_Groningen/STAGE/DATAB/copydatabase_C.db'
# Database connection
db = Database(path_db)
dict_project = get_project_ids(db)
print(len(dict_project))
```

```python
df = pd.read_sql('''SELECT project.cancer, sum_dosage_GT.donor_project_ID, 
                            sum_dosage_GT.donor_ID, sum_dosage_GT.snp_ID, 
                            snp.chr, snp.pos_start, snp.pos_end 
                    FROM project, sum_dosage_GT, snp 
                    WHERE sum_dosage_GT.snp_ID=snp.ID AND 
                              sum_dosage_GT.donor_project_ID = project.ID AND 
                              (sum_dosage_GT.GT2 = 1 OR sum_dosage_GT.GT2 = 2) AND 
                              sum_dosage_GT.total_read_count_sum >= 33;''', db.mydb_connection)
df
```

# GT0

```python
GT0 = pd.read_sql("""
                    SELECT project.cancer, sum_dosage_GT.donor_project_ID, 
                            sum_dosage_GT.donor_ID, sum_dosage_GT.snp_ID, 
                            snp.chr, snp.pos_start, snp.pos_end, COUNT(sum_dosage_GT.GT2) as GT_0
                    FROM sum_dosage_GT, snp, project
                    WHERE sum_dosage_GT.snp_ID=snp.ID AND 
                            sum_dosage_GT.donor_project_ID = project.ID AND
                            sum_dosage_GT.GT2 = 0 AND
                            sum_dosage_GT.total_read_count_sum >= 33
                    GROUP BY project.cancer, sum_dosage_GT.donor_project_ID, 
                            sum_dosage_GT.donor_ID, sum_dosage_GT.snp_ID, 
                            snp.chr, snp.pos_start, snp.pos_end
                    ORDER BY project.cancer, sum_dosage_GT.donor_project_ID, 
                            sum_dosage_GT.donor_ID, sum_dosage_GT.snp_ID, 
                            snp.chr, snp.pos_start, snp.pos_end
                    """, db.mydb_connection)
GT0
```

```python
merge_0 = df.merge(GT0, on=['cancer', 'donor_project_ID', 'donor_ID', 'snp_ID', 'chr', 'pos_start', 'pos_end'], how='outer')
merge_0
```

# GT1

```python
GT1 = pd.read_sql("""
                    SELECT project.cancer, sum_dosage_GT.donor_project_ID, 
                            sum_dosage_GT.donor_ID, sum_dosage_GT.snp_ID, 
                            snp.chr, snp.pos_start, snp.pos_end, COUNT(sum_dosage_GT.GT2) as GT_1
                    FROM sum_dosage_GT, snp, project
                    WHERE sum_dosage_GT.snp_ID=snp.ID AND 
                            sum_dosage_GT.donor_project_ID = project.ID AND
                            sum_dosage_GT.GT2 = 1 AND
                            sum_dosage_GT.total_read_count_sum >= 33
                    GROUP BY project.cancer, sum_dosage_GT.donor_project_ID, 
                            sum_dosage_GT.donor_ID, sum_dosage_GT.snp_ID, 
                            snp.chr, snp.pos_start, snp.pos_end
                    ORDER BY project.cancer, sum_dosage_GT.donor_project_ID, 
                            sum_dosage_GT.donor_ID, sum_dosage_GT.snp_ID, 
                            snp.chr, snp.pos_start, snp.pos_end
                    """, db.mydb_connection)
GT1
```

```python
merge_1 = merge_0.merge(GT1, on=['cancer', 'donor_project_ID', 'donor_ID', 'snp_ID', 'chr', 'pos_start', 'pos_end'], how='outer')
merge_1
```

# Filter data

```python
breast_cancer = merge_2.loc[merge_2['cancer'] == 'Breast']
# breast_cancer1 = breast_cancer[['chr', 'pos_start', 'pos_end', 'GT_0', 'GT_1', 'GT_2']]
# breast_cancer1.sort_values(by=['chr', 'pos_start'], inplace=True)
breast_cancer['snp'] = breast_cancer['chr'] + '_' + breast_cancer['pos_start'].map(str) + '_' + breast_cancer['pos_end'].map(str)
print(len(list(set(breast_cancer['donor_ID']))))
num_donor_breast = len(list(set(breast_cancer['donor_ID'])))

select_breast_cancer = breast_cancer[['GT_0', 'GT_1', 'GT_2', 'snp', 'snp_ID', 'chr', 'pos_start', 'pos_end']]

select_breast_cancer.sort_values(by=['snp_ID'], inplace=True)
select_breast_cancer
```

```python
count_GT_b = select_breast_cancer.groupby(by=['snp', 'snp_ID', 'chr', 'pos_start', 'pos_end']).sum()
count_GT_b
```

```python
count_GT_b.to_csv(f"D:/Hanze_Groningen/STAGE/R/GT_breast.tsv", sep='\t', encoding='utf-8')
```

```python
nonbreast_cancer = merge_2.loc[merge_2['cancer'] != 'Breast']
# nonbreast_cancer1 = nonbreast_cancer[['chr', 'pos_start', 'pos_end', 'GT_0', 'GT_1', 'GT_2']]
# nonbreast_cancer1.sort_values(by=['chr', 'pos_start'], inplace=True)
nonbreast_cancer['snp'] = nonbreast_cancer['chr'] + '_' + nonbreast_cancer['pos_start'].map(str) + '_' + nonbreast_cancer['pos_end'].map(str)
print(len(list(set(nonbreast_cancer['donor_ID']))))
num_donor_nonbreast = len(list(set(nonbreast_cancer['donor_ID'])))

select_nonbreast_cancer = nonbreast_cancer[['GT_0', 'GT_1', 'GT_2', 'snp', 'snp_ID', 'chr', 'pos_start', 'pos_end']]

select_nonbreast_cancer.sort_values(by=['snp_ID'], inplace=True)
select_nonbreast_cancer
```

```python
count_GT_nb = select_nonbreast_cancer.groupby(by=['snp', 'snp_ID', 'chr', 'pos_start', 'pos_end']).sum()
count_GT_nb
```
