```python
path_file = "D:/Hanze_Groningen/STAGE/GT2.tsv"
df = pd.read_csv(path_file, sep='\t')
df
```

```python
#286 and 1952
df['GT_b_sum'] = (df['GT_0_b'] * 0) + (df['GT_1_b'] * 1) + (df['GT_2_b'] * 2)
df['b_donors'] = df['GT_0_b'] + df['GT_1_b'] + df['GT_2_b']

df['GT_nb_sum'] = (df['GT_0_nb'] * 0) + (df['GT_1_nb'] * 1) + (df['GT_2_nb'] * 2)
df['nb_donors'] = df['GT_0_nb'] + df['GT_1_nb'] + df['GT_2_nb']

df['GT_total_sum'] = df['GT_b_sum'] + df['GT_nb_sum']
df['total_donors'] = df['b_donors'] + df['nb_donors']
df
```

```python
df['AF_breast'] = df['GT_b_sum'] / df['b_donors']
df['AF_nonbreast'] = df['GT_nb_sum'] / df['nb_donors']
df['AF_total'] = df['GT_total_sum'] / df['total_donors']
df

```
